본문 내용 및 해시태그 크롤링 함수

```python
def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html')

    #본문 내용 크롤링
    try:
        content = soup.select('div._a9zs')[0].text
    except:
        content = ''

    #해시태그 크롤링
    tags = re.findall(r'#[^\s#,\\]+', content)

    data = [content, tags]
    return data
```

데이터 전처리

```python
def clean_korean_documents(documents):
    # 텍스트 정제 (형태소 분석)
    for i, document in enumerate(documents):
        clean_words = []
        for word in okt.pos(document, stem=True):
            if word[1] in ['Noun', 'Verb', 'Adjective']: # 명사/동사/형용사만 추출
                clean_words.append(word[0])
        document = ' '.join(clean_words)
        documents[i] = document

    # 텍스트 정제 (불용어 제거)
    f = open('stopword_ko.txt', 'r', encoding='utf8')
    stopwords_ko = f.readlines()
    for i, document in enumerate(documents):
        clean_words = [] 
        for word in word_tokenize(document): 
            if word not in stopwords_ko:
                clean_words.append(word)
        document = ' '.join(clean_words)
        documents[i] = document

    return documents
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC

model_nb = MultinomialNB(alpha=1.0) # 나이브베이즈 알고리즘 기반의 분류기 생성
model_svm = SVC(kernel='linear') # 서포트벡터머신 알고리즘 기반의 분류기 생성
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3) 

model_nb.fit(x_train, y_train) 
model_svm.fit(x_train, y_train) 
```

### 빈도 기반 키워드 추출

가장 큰 tf-idf 가중치를 갖는 단어 출력

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer().fit(total) 
vocab = tfidf_vectorizer.vocabulary_ 
text_tfidf = tfidf_vectorizer.transform(total).toarray() 

tfidf_keyword = []
# 1. 문장별로 가장 큰 tfidf 가중치를 갖는 배열값의 인덱스(위치) 찾기
for sent_tfidf in text_tfidf:
  max_idx = sent_tfidf.argmax() 

  # 2. 해당 인덱스가 어떤 단어에 해당하는지 찾기
  for k,v in vocab.items(): 
    if v == max_idx:
      tfidf_keyword.append(k)
```

tf-idf 가중치가 임계값을 넘는 단어 출력

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer().fit(total)

vocab = tfidf_vectorizer.vocabulary_ 
print(vocab)  
vocab_reverse = {v:k for k,v in vocab.items()} 
print(vocab_reverse)

text_tfidf = tfidf_vectorizer.transform(total).toarray()

tfidf_keyword2 = []
# 1. 키워드의 tfidf 임계치를 설정
threshold = 0.6

# 2. 각 문장의 단어마다 tfidf 가중치가 임계치를 넘는지 if 조건문으로 판단
# 3. 단어의 ifidf 가중치가 임계치를 넘는 경우, vocab_reverse 딕셔너리를 이용하여 단어의 인덱스(위치)로 단어를 찾기
for sent_tfidf in text_tfidf:
  for idx, word_tfidf in enumerate(sent_tfidf):
    if word_tfidf > threshold: 
      tfidf_keyword2.append(vocab_reverse[idx]) 
```

```python
tfidf_keyword2 = set(tfidf_keyword2)
```
